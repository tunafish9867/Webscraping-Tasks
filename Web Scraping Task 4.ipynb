```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import pandas as pd
import os
from selenium.common.exceptions import NoSuchElementException

options = webdriver.ChromeOptions()

# Specify the desired download directory
download_dir = "C:/Users/denis/Downloads/Thesis/San Juan City Data_12262024"

options.add_experimental_option('prefs', {
    "download.default_directory": download_dir,  # Set the directory explicitly
    "download.prompt_for_download": False,       # Disable download prompt
    "download.directory_upgrade": True,          # Allow directory upgrades
    "safebrowsing.enabled": True,                # Disable safe browsing warnings
    "plugins.always_open_pdf_externally": True,  # Automatically handle PDFs
})



driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))

URL = "https://www.sanjuancity.gov.ph/SanJuanCity/legitas-web-ordinance"
driver.get(URL)

driver.maximize_window()
driver.implicitly_wait(5)

ordinance_no = []
ordinance_desc = []

# Iterate over the sections and indices
try:
    while True:
        # Loop through table rows from tr[2] to tr[16]
        for i in range(2, 17):  # Loop through rows
            try:
                # XPath for the first column (td[1])
                column1_xpath = f'//*[@id="page-container"]/section[2]/div[2]/table/tbody/tr[{i}]/td[1]'
                column1 = driver.find_element(By.XPATH, column1_xpath)
                column1_text = column1.text
                ordinance_no.append(column1_text)

                # XPath for the second column (td[2])
                column2_xpath = f'//*[@id="page-container"]/section[2]/div[2]/table/tbody/tr[{i}]/td[2]'
                column2 = driver.find_element(By.XPATH, column2_xpath)
                column2_text = column2.text
                ordinance_desc.append(column2_text)

                # Print both column texts
                print(f"Row {i}: Column 1 = {column1_text}, Column 2 = {column2_text}")
            except NoSuchElementException:
                print(f"Row {i} or its columns not found on this page.")
                continue

                    # Check if the "Next" button is present and click it
        try:
            next_button = driver.find_element(By.XPATH, '//li[@class="page-item"]/a[@class="page-link" and @rel="next"]')
            next_button.click()
            time.sleep(2)  # Wait for the next page to load
        except NoSuchElementException:
            print("No 'Next' button found. Exiting loop.")
            break
    
    df_main = pd.DataFrame({
        'Ordinance Number': ordinance_no,  # First element is Ordinance NO
        'Ordinance Desc': ordinance_desc  # Second element is Ordinance Desc
    })

    # Save the DataFrame to an Excel file


    folder = "C:/Users/denis/Downloads/Thesis/San Juan City Data_12262024"
    excel_file = os.path.join(folder, "san_juan_ordinance_details.xlsx")
    with pd.ExcelWriter(excel_file) as writer:
        df_main.to_excel(writer, sheet_name='Ordinances', index=False)
    print(f"Data saved to {excel_file}")   


finally:
    # Close the browser
    driver.quit()
```
